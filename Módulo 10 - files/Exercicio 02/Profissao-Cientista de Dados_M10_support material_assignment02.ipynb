1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
df = df.drop(columns='Unnamed: 0').set_index(keys='index', drop=True)
df.drop_duplicates(inplace=True)
df
````

````python
plt.rc('figure', figsize=(12,12))
fig, axes = plt.subplots(2,2)

sns.pointplot(x='posse_de_imovel', 
              y='renda',  
              data=df, 
              dodge=True, 
              ax=axes[0,0])

sns.pointplot(x='posse_de_veiculo', 
              y='renda', 
              data=df, 
              dodge=True, 
              ax=axes[0,1])

sns.pointplot(x='posse_de_imovel', 
              y='renda',  
              hue='posse_de_veiculo', 
              data=df, 
              dodge=True, 
              ax=axes[1,0])

sns.pointplot(x='posse_de_veiculo', 
              y='renda', 
              hue='posse_de_imovel', 
              data=df, 
              dodge=True, 
              ax=axes[1,1])

plt.show()
````

A análise dos gráficos apresentados revela que, na análise inicial, ambas as variáveis qualitativas exibem uma distinção estatisticamente significativa na renda. Contudo, destaca-se a variável posse_de_veículo por sua maior importância na previsão de renda, evidenciada pelo maior intervalo de confiança entre os grupos com e sem veículo. Em um exame mais detalhado dos gráficos subsequentes, observa-se que a variável posse_de_imóvel demonstra uma relevância menor. Isso ocorre porque a presença de veículo indica um nível de renda similar, independente da posse de imóvel, sugerindo que a posse_de_veículo é um indicador mais robusto de diferenças de renda do que a posse_de_imóvel.
