#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
crosstab = pd.crosstab(df['possui_email'], df['posse_de_veiculo'], normalize='index')
print(crosstab)
````

````python
df_cross = pd.crosstab(df['possui_email'], df['posse_de_veiculo'], values=df['mau'], aggfunc='mean')
df_cross['mau'] = df['mau'].mean()

ax = df_cross.plot.bar(width=0.5)
ax.set_xlabel('Possui E-mail')
ax.set_ylabel('Proporção')
ax.set_title('Proporção de mau de acordo com posse de veículo e e-mail')

pos_veiculo_labels = ['Não possui', 'Possui']
pos_email_labels = ['Não possui', 'Possui']
ax.set_xticklabels(pos_email_labels)
ax.legend(pos_veiculo_labels, title='Posse de Veículo', loc='center left', bbox_to_anchor=(1.0, 0.5))

ax.set_ylim(top=0.06)
ax.tick_params(axis='x', labelsize=0)

plt.show()
````

Analisando a tabela cruzada resultante, podemos ver que a proporção de mau=1 é ligeiramente maior para pessoas que possuem um veículo em comparação com aquelas que não possuem um veículo, independentemente de possuírem ou não um e-mail. Por exemplo, a proporção de mau=1 para pessoas que não possuem um e-mail e possuem um veículo é de 37,50%, enquanto a proporção para pessoas que não possuem um e-mail e não possuem um veículo é de 62,50%. Similarmente, a proporção de mau=1 para pessoas que possuem um e-mail e um veículo é de 41,06%, enquanto a proporção para pessoas que possuem um e-mail e não possuem um veículo é de 58,94%.

Com base nessa tabela cruzada, podemos concluir que a variável posse_de_veiculo pode ser mais poderosa para prever a probabilidade de mau=1 do que a variável possui_email.
