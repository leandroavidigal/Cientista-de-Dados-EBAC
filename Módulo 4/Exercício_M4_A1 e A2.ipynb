#### 1)  Crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na tempreatura global, A NASA disponibiliza dados de diferenças de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Crie uma séries do Pandas a partir de uma lista com esses dados.

#### 4) Transforme o ndarray abaixo em um dataframe. 
O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

Explicando cada valor da função para criação da tabela.
- O primeiro parâmetro (100) representa a média da distribuição normal.
- O segundo parâmetro (10) representa o desvio padrão da distribuição normal.
- O terceiro parâmetro ((20,3)) representa a forma (shape) do nd-array que será gerado. Neste caso, será um array com 20 linhas e 3 colunas.

```python
arr = np.random.normal(100, 10, (20,3))

df = pd.DataFrame(arr) # Cria o DataFrame a partir do nd-array
print(df)
```

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#definindo os nomes das colunas
df = pd.DataFrame(arr, columns=["x1", "x2", "x3"])

#definindo os rótulos dos índices das linhas
df = df.rename(index=lambda x: x + 1)
print(df)
```

#### 6) Crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media".

```python
#calculando e criando uma nova coluna com o valor das médias
df = df.assign(media=df.mean(axis=1))
print(df)
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
#calculando o logaritmo natural da média
df = df.assign(log_med=np.log(df["media"]))
print(df)
```
